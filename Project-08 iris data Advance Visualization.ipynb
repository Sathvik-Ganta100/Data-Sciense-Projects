## 7. Combining Box and Strip Plots

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.boxplot(data = iris, x = 'Species', y = 'SepalLengthCm')
fig=sns.stripplot(data = iris, x = 'Species', y = 'SepalLengthCm', hue = 'Species')
```

```python
colors = ['green', 'red', 'yellow']

sns.boxplot(data = iris, x = "Species", y = "PetalLengthCm", palette = colors)
sns.stripplot(data = iris, x = "Species", y = "PetalLengthCm", jitter = True, hue = 'Species')



plt.show()
```

## 17. Boxen plot

```python
sns.set_style('whitegrid')

fig=plt.gcf()
fig.set_size_inches(7, 5)
fig=sns.boxenplot(data = iris, x = 'Species', y = 'SepalLengthCm')
```

## 18. KDE Plot

```python
sub = iris[iris['Species'] == 'Iris-setosa']

sns.kdeplot(data=sub , x='SepalLengthCm', y='SepalWidthCm', cmap="plasma", shade=True, shade_lowest=False, cbar=True)

plt.title('Iris-setosa')
plt.xlabel('Sepal Length Cm')
plt.ylabel('Sepal Width Cm')

plt.show()
```

## 19. Dashboard

```python
sns.set_style('darkgrid')
f,axes=plt.subplots(2,2,figsize=(10,10))

k1=sns.boxplot(data=iris,x="Species", y="PetalLengthCm",ax=axes[0,0])
k2=sns.violinplot(data=iris,x='Species',y='PetalLengthCm',ax=axes[0,1])
k3=sns.stripplot(data=iris,x='Species',y='SepalLengthCm',jitter=True,edgecolor='gray',palette='winter',orient='v',ax=axes[1,0])
axes[1,1].hist(iris.PetalLengthCm,bins=10)

plt.show()
```

## 20. Stacked Histogram

```python
iris['Species'] = iris['Species'].astype('category')

iris.info()
```

```python
list1=list()
mylabels=list()
for gen in iris.Species.cat.categories:
    list1.append(iris[iris.Species==gen].SepalLengthCm)
    mylabels.append(gen)
    
h=plt.hist(list1,bins=30,stacked=True,rwidth=1,label=mylabels)

plt.legend()
plt.show()
```
