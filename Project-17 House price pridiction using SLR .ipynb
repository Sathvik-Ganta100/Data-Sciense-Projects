# House price pridiction using SLR

```python
#Import libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# importing dataset and spliting into dv and iv

ds = pd.read_csv(r'D:\1. Professionall\Data Science\08-11-2023\House_data.csv')

space = ds['sqft_living']
price = ds['price']

X = np.array(space).reshape(-1,1)
y = np.array(price)
```

```python
# spliting data into test and train data

from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X,y,test_size=0.2, random_state=0)
```

```python
# Built regression model

from sklearn.linear_model import LinearRegression 
regressor = LinearRegression()
regressor.fit(Xtrain, ytrain)
```

```python
# Test the model and prepare future prediction table

ypred = regressor.predict(Xtest)
```

```python
# visulize training data set

plt.scatter(Xtrain, ytrain, color= 'red')
plt.plot(Xtrain, regressor.predict(Xtrain), color = 'blue')
plt.title ("Visuals for Training Dataset")
plt.xlabel("Space")
plt.ylabel("Price")
plt.show()
```

```python
# visualizing test dataset 

plt.scatter(Xtest, ytest, color= 'red')
plt.plot(Xtrain, regressor.predict(Xtrain), color = 'blue')
plt.title("Visuals for Test DataSet")
plt.xlabel("Space")
plt.ylabel("Price")
plt.show()
```

```python
# To get value of slope
m = regressor.coef_
m
```

```python
# To get intercept or constant  
c = regressor.intercept_
c
```
