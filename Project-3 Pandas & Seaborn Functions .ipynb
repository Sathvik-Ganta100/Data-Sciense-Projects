# Seaborn         

* Introduction to seaborn     
* seaborn is very powerfull visualizatio(STATISTIC VISULAIZATION) pkg in python
* seaborn are used for advance visualization e.x ===> distribution plot, line plot

```python
import matplotlib.pyplot as plt   # visulaiztion
import seaborn as sns             # distribution visualtion

%matplotlib inline
plt.rcParams['figure.figsize'] = 10,5    # increase fig size to height 10 and width 5

import warnings
warnings.filterwarnings('ignore')  # ignore OS warnings
```

```python
cd.head()
```

```python
cd['InternetUsers']
```

```python
# Distributions:

vis1 = sns.distplot(cd['InternetUsers'])   # UNIVARIATE ANALYSIS - statistics 
                                           # .distplto() ---> distribution plot
```

```python
vis1 = sns.distplot(cd['InternetUsers'], bins = 20)
```

```python
sns.boxplot(cd)  # direct plots a candle chart for numerical data
```

```python
sns.boxplot(cd.BirthRate)  # if we want a specific candle chart for a numeric data
                           # This process is not possible for categorical data
```

```python
vis2 = sns.boxplot(data = cd, x = 'IncomeGroup', y = 'BirthRate')  # BI-VARIATE ANALYSIS 
                                                                    # By this process we can plot candle chart b/w 
                                                                    # numerical and categorical data 
                                                                    # NOTE: plot b/w two categorical is not possible
 # ANOMALY DETECTION or OUTLIER ---> outlier is the point which significanely far away form other data point?
 #             Here Diamond point is a outliers
```

```python
vis2 = sns.boxplot(data = cd, y = 'IncomeGroup', x = 'BirthRate')  # Above graph iv horizontal format 
                                                                  #    (x & y position chainged)
```

```python
vis3 = sns.lmplot(data = cd, x = 'InternetUsers', y = 'BirthRate')    # linear model plot
                                                                      # can only plot b/w numerical data 
```

```python
vis3 = sns.lmplot(data = cd, x = 'InternetUsers', y = 'BirthRate',fit_reg=False)  # removes line
```

```python
vis3 = sns.lmplot(data = cd, x = 'InternetUsers', y = 'BirthRate',fit_reg = False, hue = 'IncomeGroup')  
                                                                      # To differentiate based on third group
```

```python
vis3 = sns.lmplot(data = cd, x = 'InternetUsers', y = 'BirthRate',fit_reg = False, hue = 'IncomeGroup', height = 7) 
                                                                    
     # To increase size
```

```python
vis3 = sns.lmplot(data = cd, x = 'InternetUsers', y = 'BirthRate',fit_reg = True, hue = 'IncomeGroup') 
```

```python
vis3 = sns.lmplot(data = cd, x = 'InternetUsers', y = 'BirthRate',fit_reg = True, hue = 'IncomeGroup', height=10) 
```
