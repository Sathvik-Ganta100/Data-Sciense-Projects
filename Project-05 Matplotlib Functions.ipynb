## 1. Import Matplotlib

```python
import matplotlib.pyplot as plt         # to import Matplotlib 

import numpy as np                   # import dependencies
import pandas as pd

```

## 2. Plotting from a Jupyter notebook

```python
%matplotlib inline   

# %matplotlib inline ---> It sets up the notebook environment to render plots as static image

x1 = np.linspace(0, 10, 100)


# create a plot figure
fig = plt.figure()

plt.plot(x1, np.sin(x1), '-')
plt.plot(x1, np.cos(x1), '--')
```

```python
plt.plot(x1, np.tan(x1), '-.')
```

## 4. Pyplot API

* This is really helpful for interactive plotting
* We can issue a command and see the result immediately
* Not suitable for more complicated cases

```python
# create a plot figure
plt.figure()


# create the first of two panels and set current axis
plt.subplot(2, 1, 1)   # (rows, columns, panel number)
plt.plot(x1, np.sin(x1))


# create the second of two panels and set current axis
plt.subplot(2, 1, 2)   # (rows, columns, panel number)
plt.plot(x1, np.cos(x1))


print(plt.gcf())   # get current figure information
print(plt.gca())   # get current axis information
```

### Visualization with Pyplot

```python
plt.plot([1, 2, 3, 4, 5])    # by default takes as Y axis coordinates and X axis coordinates start from 0
plt.ylabel('Numbers')        # creates lable on Y axis

plt.show()    # is a function in the matplotlib.pyplot module that is used to display the figure
```

### plot() - A versatile command

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])   # plt,plot([X],[Y])---> to plot X versus Y, we can issue the following command

plt.show()
```

### State-machine interface

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()

plt.show()
```

### Formatting the style of plot

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')   # ro ---> r - red color | o - circle marker
plt.axis([0, 6, 0, 20])        # is used to set the axis limits for the X and Y axes of a plot (here limit on Y axis)

plt.show()
```

### Working with NumPy arrays

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')  # giving multiple datasets in a single plot comand

plt.show()
```

## 5. Object-Oriented API

* It is available for more complex plotting situations
* It allows us to exercise more control over the figure
* Suitable for more complicated cases

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(3)


# Call plot() method on the appropriate object
ax[0].plot(x1, np.sin(x1), 'g-')
ax[1].plot(x1, np.cos(x1), 'b-')
ax[2].plot(x1, np.tan(x1), 'r-.')
```

### Objects and Reference

```python
fig = plt.figure()

x2 = np.linspace(0, 5, 100)
y2 = x2 ** 2

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])   # adds a new set of axes to the existing figure - [left, bottom, width, height]

axes.plot(x2, y2, 'r')

axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('title')
```

## 16. Error Bar Chart

```python
x9 = np.arange(0, 4, 0.2)

y9 = np.exp(-x9)

e1 = 0.1 * np.abs(np.random.randn(len(y9)))

plt.errorbar(x9, y9, yerr = e1, fmt = '.-')

plt.show()
```

 ## 17. Stacked Bar Chart

```python
A = [15., 30., 45., 22.]

B = [15., 25., 50., 20.]

z2 = range(4)

plt.bar(z2, A, color = 'b')
plt.bar(z2, B, color = 'r', bottom = A)

plt.show()
```

## 24. Handling axes

```python
x15 = np.arange(1, 5)

plt.plot(x15, x15*1.5, x15, x15*3.0, x15, x15/3.0)

plt.axis()   # shows the current axis limits values

plt.axis([0, 5, -1, 13])  # 0 to 5 is X axis limit & -1 to 13 is Y axis limit

plt.show()
```

```python
# We can control the limits for each axis separately using the xlim() and ylim() functions

x15 = np.arange(1, 5)

plt.plot(x15, x15*1.5, x15, x15*3.0, x15, x15/3.0)

plt.xlim([1.0, 4.0])

plt.ylim([0.0, 12.0])
```

## 25. Handling X and Y ticks

## 27. Adding a legend

```python
x15 = np.arange(1, 5)

fig, ax = plt.subplots()

ax.plot(x15, x15*1.5)
ax.plot(x15, x15*3.0)
ax.plot(x15, x15/3.0)

ax.legend(['Normal','Fast','Slow'])
```

```python
# The above method follows the MATLAB API. 
# It is prone to errors and unflexible if curves are added to or removed from the plot

# A better method is to use the label keyword argument when plots are added to the figure (Below method)
```

```python
x15 = np.arange(1, 5)

fig, ax = plt.subplots()

ax.plot(x15, x15*1.5, label='Normal')
ax.plot(x15, x15*3.0, label='Fast')
ax.plot(x15, x15/3.0, label='Slow')

ax.legend()
```
